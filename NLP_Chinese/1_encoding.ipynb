## 編碼 chardet library

```python
# 直接指定編碼 utf-8 (但實際檔案並不是, 所以會有錯！)
filename = 'redmansion_big5.txt'
with open(filename, 'r', encoding='utf-8') as f:
  article = f.read()

print(article)
```

```python
import chardet

filename_big5 = 'redmansion_big5.txt'

# 先偵測檔案編碼
with open(filename_big5, 'rb') as f:  # (b)原始內容讀出
  content = f.read()
  en = chardet.detect(content)

print(en)
print(en['encoding'])
```

```python
filename_utf8 = 'redmansion_big5_utf8.txt'

# 寫入檔案，使用 utf-8 編碼
with open(filename_utf8, 'w', encoding='utf-8') as f:
  content_utf8 = content.decode(en['encoding'], errors='ignore')  # ignore 解不出來的字就跳過
  f.write(content_utf8)

print(content_utf8[-1000:])
```

## 讀入簡體中文


```python
# 測試：查看 GBK　內容格式
filename_gbk = 'redmansion_gbk.txt'

# 先偵測檔案編碼
with open(filename_gbk, 'rb') as f:
  content = f.read()
  en = chardet.detect(content)
  print(en)

content_utf8 = content.decode(en['encoding'], errors='ignore')
print("Before:", content_utf8[-1000:])
```

## opencc 程式庫 （轉換成繁體）

```python
!pip install opencc-python-reimplemented
```

```python
from opencc import OpenCC

cc = OpenCC('s2tw')  # 台灣標準
converted = cc.convert(content_utf8)
print("After:", converted[-1000:])

# 寫入檔案，使用 utf-8 編碼
filename_tw_utf8 = 'remansion_tw_utf-8.txt'
with open(filename_tw_utf8, 'w', encoding='utf-8') as f:
  f.write(converted)
```
