```python
# 載入相關套件
import os
import tempfile

from matplotlib import pyplot as plt
import numpy as np
import tensorflow as tf
```

## 載入預訓練的TF模型以及儲存到本地

```python
pretrained_model = tf.keras.applications.MobileNet()
```

```python
model_save_path = os.path.join(os.getcwd(), "model/")
tf.saved_model.save(pretrained_model, model_save_path)
```

## 轉換模型（TF→TF.JS）

```python
output_dir = '/content/tfjs_model'
```

```python
!tensorflowjs_converter --input_format=tf_saved_model --output_format=tfjs_graph_model $model_save_path $output_dir
```

# 載入TF.JS模型以及預測

```python
from IPython.display import HTML
```

```python
html_code = HTML('''
<!DOCTYPE html>
<html>
<head>
    <title>TensorFlow.js Example</title>
    <script src="https://cdn.jsdelivr.net/npm/@tensorflow/tfjs@3.8.0"></script>
</head>
<body>
  <div id="output"></div>
  <script>
    const a = tf.tensor([[1, 2], [3, 4]], [2, 2], 'int32');
    console.log('shape:', a.shape);
    console.log('dtype', a.dtype);
    document.querySelector("#output").appendChild(document.createTextNode('shape:'+a.shape));
    document.querySelector("#output").appendChild(document.createTextNode('dtype:'+a.dtype));
  </script>
</body>
</html>
''')

html_code
```

```python
html_code = HTML('''
<!DOCTYPE html>
<html>
<head>
    <title>TensorFlow.js Example</title>
    <script src="https://cdn.jsdelivr.net/npm/@tensorflow/tfjs@3.8.0"></script>
</head>
<body>
  <script>

  async function run() {
    const MODEL_URL = 'localstorage://content/tfjs_model/model.json';
    const model = await tf.loadGraphModel(MODEL_URL);
    console.log(model);
    const cat = document.getElementById('cat');
    const prediction = await model.predict(tf.browser.fromPixels(cat));
    console.log(prediction);
  }

  run();

</script>
</body>
</html>
''')

html_code
```
